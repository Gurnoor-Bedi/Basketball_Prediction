```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Create a DataFrame with more data
data = {
    'Week': list(range(1, 201)),
    'Free Throws Attempted': [random.randint(40, 70) for _ in range(200)],
    'Free Throws Made': [random.randint(30, 60) for _ in range(200)],
    '2-Pointers Made': [random.randint(10, 30) for _ in range(200)],
    '3-Pointers Made': [random.randint(5, 20) for _ in range(200)],
    'Passes Made': [random.randint(60, 100) for _ in range(200)]
}

df = pd.DataFrame(data)

# Calculate Free Throw Percentage
df['Free Throw Percentage (%)'] = (df['Free Throws Made'] / df['Free Throws Attempted']) * 100

# Calculate Average Free Throw Percentage
average_ft_percentage = df['Free Throw Percentage (%)'].mean()

# Display the updated DataFrame
df

```

```python
print("\nAverage Free Throw Percentage: {:.2f}%".format(average_ft_percentage))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Free Throws Attempted'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Free Throws Attempted')
plt.ylabel('Frequency')
plt.title('Distribution of Free Throws Attempted')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Week'], df['2-Pointers Made'], marker='o', label='2-Pointers Made')
plt.plot(df['Week'], df['3-Pointers Made'], marker='o', label='3-Pointers Made')
plt.xlabel('Week')
plt.ylabel('Number of Shots Made')
plt.title('2-Pointers and 3-Pointers Made Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Week'], df['Passes Made'], marker='o', color='purple', label='Passes Made')
plt.xlabel('Week')
plt.ylabel('Number of Passes Made')
plt.title('Passes Made Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to your data
X = df[['Week']].values
y = df['Free Throw Percentage (%)'].values
model.fit(X, y)

# Predict the free throw percentage for the next week (Week 21)
next_week = 21
predicted_percentage = model.predict(np.array([[next_week]]))

print("Predicted Free Throw Percentage for Week {}: {:.2f}%".format(next_week, predicted_percentage[0]))
```
